```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
```

```python
df=pd.read_csv("canada_per_capita_income.csv")
df.head()
```

```python
plt.title("Year Vs PCI($)")
plt.xlabel("year")
plt.ylabel('per capita income(US$)')
plt.scatter(df.year,df['per capita income (US$)'],color="red",marker='+')
plt.show()
```

```python
year=df.drop('per capita income (US$)',axis=1)
year.head()
```

```python
pci=df['per capita income (US$)']
pci.head()
```

```python
reg=linear_model.LinearRegression()
reg.fit(year,pci)
```

```python
reg.predict([[2020]])
```

```python
plt.title("Linear Regressio0n (Year Vs Pci($))")
plt.xlabel("year")
plt.ylabel('per capita income (US$)')
plt.plot(year,reg.predict(year),color='red')
plt.show()
```

```python
for i in range(2020,2031):
    print(f"{i}:{reg.predict([[i]])}")
```
