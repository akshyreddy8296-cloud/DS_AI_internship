<h2>Phase 1: The Detective Work (Setup & Inspection)</h2>

```python
import pandas as pd
```

```python
df=pd.read_csv("customer_analytics.csv")
print(df.head())
print(df.info())
print(df.describe())
```

```python
print("Insights:")
print("Dataset Name: customer_analytics.csv")
rows, cols = df.shape
print(f"Dataset Shape: {rows} rows, {cols} columns")
print(f"Data types:\n {df.dtypes}")
print("\nMissing Values (per column):")
print(df.isnull().sum())
print("\nBasic Statistics (Numerical Columns):")
print(df.describe())
print("\nCategorical Summary:")
print(df.describe(include='object'))
print("\nDataset Info:")
df.info()
```

<h2>Phase 2: The Cleanup (Data Preprocessing)</h2>

```python
print("Missing Values:")
print(df.isnull().sum())

# Imputation (safe method)
df["Education"] = df["Education"].fillna(df["Education"].mode()[0])
df["AnnualIncome"] = df["AnnualIncome"].fillna(df["AnnualIncome"].median())

print("\nMissing Values After Imputation:")
print(df.isnull().sum())

# Check duplicates
duplicates = df.duplicated().sum()
print(f"\nDuplicate Rows: {duplicates}")

df = df.drop_duplicates()

print(f"Dataset Shape After Cleaning: {df.shape}")
```

<h3>Missing Values:-</h3>

The dataset contains 12 missing values in both Education and AnnualIncome.
Since this dataset has only 255 rows, dropping rows would remove valuable information.
Therefore, imputation was preferred over deletion.

<h3>Imputation Strategy</h3>

Education (Categorical):
Filled with the mode (most frequent value) because it preserves category distribution.

AnnualIncome (Numerical):
Filled with the median instead of the mean because income data often contains outliers.
Median is more robust and prevents distortion.

<h3>Duplicates</h3>

Duplicate rows were checked using .duplicated().

5 duplicates found were removed to ensure each row represents a unique customer.

<h2>Phase 3: Deep Dive (Univariate & Bivariate Analysis)</h2>

```python
import matplotlib.pyplot as plt
import seaborn as sns

# ---------- Univariate Plot 1: Age Distribution ----------
plt.figure()
plt.hist(df["Age"], bins=10)
plt.title("Distribution of Customer Age")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

<h3>Insights:</h3><h4>Most customers fall between ages 40 and 45, indicating the dataset mainly represents working-age adults. Very few customers are at the extreme young or older ends.</h4>

```python
# ---------- Univariate Plot 2: Annual Income ----------
plt.figure()
plt.hist(df["AnnualIncome"], bins=10)
plt.title("Distribution of Annual Income")
plt.xlabel("Income")
plt.ylabel("Count")
plt.show()
```

<h3>Insights:</h3><h4>Income distribution is spread widely with some high-income outliers. This suggests customers come from diverse economic backgrounds. Also, majority of customers have low income.</h4>

```python
# ---------- Univariate Plot 3: Gender Count ----------
plt.figure()
df["Gender"].value_counts().plot(kind="bar")
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Number of Customers")
plt.show()
```

<h3>Insights:</h3>
<h4>Gender Ratio is almost balanced. Male customer count is slightly higher. This shows that both male and female customer equally participate in purchase.</h4>

```python
# ---------- Bivariate Plot 1: Income vs Spending ----------
plt.figure()
plt.scatter(df["AnnualIncome"], df["SpendingScore"])
plt.title("Annual Income vs Spending Score")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.show()
```

<h3>Insights:</h3>
<h4>There is no strong linear relationship between income and spending score. All high-income customers spend less, while low-income customers spend in varying patterns; indicating that income alone does not determine spending behavior.</h4>

```python
# ---------- Bivariate Plot 2: Spending by Marital Status ----------
plt.figure()
sns.boxplot(x="MaritalStatus", y="SpendingScore", data=df)
plt.title("Spending Score by Marital Status")
plt.show()
```

<h3>Insights:</h3>
<h4>Single customers have higher upper bound of spending. But, There is 1 outlier present in Married customer's Spending Score that shows excessive spending.</h4>

<h2>Phase 4: Big Picture (Multivariate & Storytelling)</h2>

```python
# Select numerical columns
num_df = df.select_dtypes(include=["int64", "float64"])

# Correlation matrix
corr = num_df.corr()

plt.figure()
sns.heatmap(corr, annot=True)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

<h3>Insights:</h3>
<h4>The heatmap reveals relationships between numerical variables. Strong positive or negative correlations indicate variables that move together, which can guide predictive modeling or feature selection.</h4>
